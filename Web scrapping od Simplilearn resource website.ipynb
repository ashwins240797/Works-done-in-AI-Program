```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen

# define url
url = 'http://www.simplilearn.com/resources'

# access result through requests objects
webpage = urlopen(url)



soup = BeautifulSoup(webpage,'html.parser')
webpage.close()
soup.contents
```

```python
# print the contents in good format
new_soup = soup.prettify()
print(new_soup)
type(soup)
```

```python
# view the head
print(soup.head)
```

```python
# view the title
print(soup.title)
```

```python
# printing all href links
for href in soup.findAll('a',href=True):
    print(href['href'])
```

```python
# view all headers name
for article in soup.findAll("h2"):
    print(article.string)
```

```python
# view all headers topic  name
for article_topic in soup.findAll("h4"):
    print(article_topic.string)
```
