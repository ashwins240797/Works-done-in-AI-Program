```python
import pandas as pd
import numpy as np

df = pd.read_csv("D://M_AI//Machine Learning//Data Preprocessing//Salaries.csv")
df
```

```python
# Total Salary cost of a company in each year
df.groupby(df.Year).TotalPay.sum()
```

```python
Total_pay_2011 = df.groupby(df.Year).TotalPay.sum()[2011]
Total_pay_2011
```

```python
Total_pay_2014 = df.groupby(df.Year).TotalPay.sum()[2014]
Total_pay_2014
```

```python
# increased total pay
increased_total_pay = Total_pay_2014 - Total_pay_2011
in_percentage = (Total_pay_2014 - Total_pay_2011)*100/Total_pay_2011

print("Thee increased total pay is %f"%increased_total_pay)
print("The increased percentage from 2011 to 2014 is %.3f:"%in_percentage )
```

```python
max_pay = df.groupby(df['Year']).TotalPay.max()
max_pay
```
