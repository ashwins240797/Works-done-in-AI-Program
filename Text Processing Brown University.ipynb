```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
with open('D://M_AI//Machine Learning//Demo Datasets//Lesson 11//brown_corpus_ca10.txt','r') as myfile:
    data = myfile.read().replace('\n', '')
```

```python
data2 = data.replace('/','')
```

```python
for i,line in enumerate(data2.split('\n')):
    if i>10:
        break
    print(str(i) + ':\t' + line)    
```

```python
# perform tokenization
from nltk import sent_tokenize, word_tokenize
sent_tokenize(data2)
```

```python
# word tokenization
for sent in sent_tokenize(data2):
    print(word_tokenize(sent))
```

```python
# removing stop words
from nltk.corpus import stopwords
import string

clean_txt = [ word for word in word_tokenize(data2) if word.lower() not in stopwords.words('english')]
clean_txt = list(map(str.lower,clean_txt))
print(clean_txt)
```

```python
# removing stopwords with punctuation
no_punc = [word for word in clean_txt if word not in string.punctuation]
no_punc = list(map(str.lower,no_punc))
print(no_punc)
```

```python
# stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for word in no_punc:
    print(ps.stem(word))
```

```python
# lemmetization
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

for word in no_punc:
    print(wnl.lemmatize(word))
```

```python
# POS Tagging
tagged = nltk.pos_tag(no_punc)
print(tagged)
```

```python
# Named Entity Recognition (NER)
sentences = nltk.sent_tokenize(data2)
tokenized_sent = [nltk.word_tokenize(sentence) for sentence in sentences]
tagged_sent = [nltk.pos_tag(sentence) for sentence in tokenized_sent]
chunked_sentences = nltk.ne_chunk_sents(tagged_sent,binary=True)

def extract_entity_names(t):
    entity_names =[]
    if hasattr(t,'label')and t.label:
        if t.label() =='NE':
            entity_names.append(' '.join([child[0]for child in t]))
        else:
            for child in t:
                entity_names.extend(extract_entity_names(child))
    return entity_names 

print(set(entity_names))
```

```python
# extract the entity names by iterating over each tree in chunked_sentences
entity_names = []
for tree in chunked_sentences:
    entity_names.extend(extract_entity_names(tree))

print(set(entity_names))
```
