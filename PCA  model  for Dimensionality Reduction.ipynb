```python


from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

n_samples = 20
random_state = 20

# generarte dataset with 10 features
X,y = make_blobs(n_samples=n_samples, n_features=10, random_state=None)
X.shape
```

```python
#define the PCA estimator with number of reduced components
pca = PCA(n_components = 3)

#fit the data into the estimator
pca.fit(X)
pca.explained_variance_ratio_
```

```python
# Transform the fitted data using transform method for dimension reduction
pca_reduced = pca.transform(X)
pca_reduced.shape
```
