```python
import matplotlib.pyplot as plt
%matplotlib inline

# cause of death
Cause = 'Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others'

# Percentage of people died due to respective cause
percentile = ['62','5','4','2','1','26']

#explode the one with highest percentile
ex = (0.05,0,0,0,0,0)

# plot the pie chart
plt.figure(figsize=(10,10))

# draw the pie and set the parameters
plt.pie(percentile,labels=Cause,explode=ex,autopct='%1.1f%%',startangle=70)
plt.title('Ohio State - Analysing cause of death')
plt.show()

```
