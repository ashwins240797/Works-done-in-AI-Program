```python
import nltk
from nltk.corpus import stopwords
import string
```

```python
filename = 'D://M_AI//Natural Language Processing//Datasets//FP.txt'
file = open(filename,'rt')
text = file.read()
print(text)
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)
```

```python
#removing the punctuations
no_punc = [word for word in tokens if word not in string.punctuation]
#no_punc =''.join(no_punc)

no_punc
```

```python
#removing stopwords
no_stop = [word for word in no_punc if word.lower() not in stopwords.words('english')]
no_stop
```

```python
#Finding the root words using Porter Stemming
from nltk.stem import PorterStemmer
porter = PorterStemmer()
stemmed =[porter.stem(word) for word in no_stop]
stemmed[:100]
```
